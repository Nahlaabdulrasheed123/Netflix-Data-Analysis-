```python

top_countries = df['country'].value_counts().head(11)
top_countries=top_countries.drop(index="unknown")

top_countries.plot(kind='bar',figsize=(10,5))
plt.title("Top 10 content producing countries")
plt.show()
```

This histogram highlights the top 10 countries contributing content to Netflix.It shows a clear imbalance,where a few countries-especially the United States and India-dominate production.The visualization reveals how global content is concentrated in limited regions.

```python
genres = df['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(y=top_genres.index, x=top_genres.values)
plt.title("Top Genres on Netflix")
plt.show()
```

The graph highlights the top 10 netflix genres showing popular categories like movies,dramas,comedies dominating the platform.

The graph shows the most Netflix content is rated TV-MA and TV-14 highlighting a focus on mature and teen audiences.

The graph shows a rapid rise in Netflix content over the years.

```python
movies = df[(df['type'] == 'Movie') & (df['unit'] == 'min')].copy()

movies['duration'] = pd.to_numeric(movies['duration'], errors='coerce')
movies = movies.dropna(subset=['duration', 'country'])

movies = movies.assign(country=movies['country'].str.split(', ')).explode('country')

top_countries = movies['country'].value_counts().head(10).index
filtered = movies[movies['country'].isin(top_countries)]


sns.barplot(data=filtered, x='duration', y='country', estimator='mean')

plt.title("Average Movie Duration by Country")
plt.show()
```

```python
pivot = df.pivot_table(index='month', columns='year', aggfunc='size')

sns.heatmap(pivot, cmap='coolwarm')

plt.title("Content Release Heatmap")
plt.show()
```

```python
movies = df[df['type'] == 'Movie']

movies['duration'] = pd.to_numeric(movies['duration'], errors='coerce')

sns.histplot(movies['duration'], bins=30, kde=True)

plt.title("Movie Duration Distribution")
plt.xlabel("Minutes")
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)
top_directors=top_directors.drop(index="unknown")

sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title("Top Directors on Netflix")
plt.show()
```

```python
df_cast = df['cast'].dropna()

actors = df_cast.str.split(', ').explode()

top_actors = actors.value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_actors.values, y=top_actors.index, palette='viridis')

plt.title("Top 10 Actors on Netflix")
plt.xlabel("Number of Appearances")
plt.ylabel("Actors")
plt.show()
```

```python
actors = df[['cast','type']].dropna()
actors['cast'] = actors['cast'].str.split(', ')
actors = actors.explode('cast')

top_actors = actors['cast'].value_counts().head(10).index
df_a = actors[actors['cast'].isin(top_actors)]

plt.figure(figsize=(12,6))
sns.countplot(data=df_a, y='cast', hue='type')
plt.title("Top Actors vs Type Relationship")
plt.show()
```

```python
df_india = df[df['country'].str.contains('India', na=False)]

plt.figure(figsize=(6,5))
sns.countplot(data=df_india, x='type', palette='Set2')
plt.title("India: Movies vs TV Shows")
plt.show()
```

```python
actors = df_india['cast'].dropna().str.split(', ').explode()

top_actors = actors.value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_actors.values, y=top_actors.index, palette='magma')
plt.title("Top Indian Actors on Netflix")
plt.xlabel("Appearances")
plt.show()
```

```python
df_india['date_added'] = pd.to_datetime(df_india['date_added'], errors='coerce')
df_india['year_added'] = df_india['date_added'].dt.year

df_india['year_added'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("India Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
